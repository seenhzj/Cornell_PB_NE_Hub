```python
from math import sin, cos, sqrt, atan2, radians


def cal_distance(lat1,lat2,lon1,lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    # change km to mile
    distance = R * c / 1.609
    
    return distance
```

```python
#facility facility distance
dist = pd.DataFrame()

initials=[]
finals=[]
distance=[]

k=0
for i in range(len(df_locations)):
    for j in range(len(df_locations)):
        distance.append(cal_distance(float(df_locations['lat'][i]),float(df_locations['lat'][j]),float(df_locations['lng'][i]),float(df_locations['lng'][j])))
        initials.append(df_locations['ID'][i])
        finals.append(df_locations['ID'][j])

dist['initial']=initials
dist['final']=finals
dist['distance']=distance


dist
```
