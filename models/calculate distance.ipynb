```python
from math import sin, cos, sqrt, atan2, radians


def cal_distance(lat1,lat2,lon1,lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    # change km to mile
    distance = R * c / 1.609
    
    return distance
```

```python
#facility facility distance
dist = np.zeros(shape=(len(df_locations)*len(df_locations),3))

k=0
for i in range(len(df_locations)):
    for j in range(len(df_locations)):
        dist[k][2] = cal_distance(df_locations['lat'][i],df_locations['lat'][j],df_locations['lng'][i],df_locations['lng'][j])
        dist[k][0]=df_locations['ID'][i]
        dist[k][1]=df_locations['ID'][j]
        k+=1
dist = pd.DataFrame(dist,columns=['initial','destination','distance'])

dist
```

```python
f= open("distance.txt","w+")
for i in range(len(dist)):
     f.write("%d " % (dist['initial'][i]))
     f.write("%d " % (dist['destination'][i]))
     f.write("%d\r\n" % (dist['distance'][i]))
f.close() 
```
