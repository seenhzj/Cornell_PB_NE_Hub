## Visualization prepare

### deliver

```python
result = engine.execute('select originfacilityzip,origin_major_city,origin_lat,origin_lng,sum(count) from deliver_package_info group by origin_major_city')
origin_df = pd.DataFrame(result.fetchall(),columns=['facilityzip','major_city','lat','lng','sum'])

```

```python
for i in range(len(origin_df)):
    if np.isnan(origin_df['lat'][i]):
        origin_df = origin_df.drop(i)
```

```python
origin_df['type']=1
```

```python
result = engine.execute('select destinationfacilityzip,dest_major_city,dest_lat,dest_lng,sum(count) from deliver_package_info group by dest_major_city')
dest_df = pd.DataFrame(result.fetchall(),columns=['facilityzip','major_city','lat','lng','sum'])

```

```python
for i in range(len(dest_df)):
    if np.isnan(dest_df['lat'][i]):
        dest_df = dest_df.drop(i)
```

```python
dest_df['type']=2
dest_df
```

```python
deliver_df=origin_df.append(dest_df)
```
