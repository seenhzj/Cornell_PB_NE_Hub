```python
from folium.features import CustomIcon
import os


path = os.getcwd()
hub_image = path+'/hub_icon.png'

finalhub_image = path+'/finalhub_icon.png'

## Customize the facility icon
def add_facility_marker(df,name,m,finallist): 
    path = os.getcwd()
    hub_image = path+'/hub_icon.png'
    finalhub_image = path+'/finalhub_icon.png'
    
    for i in range(len(df)):
#         if name=='final facility':
        if df.iloc[i]['ID'] in finallist:
            icon = CustomIcon(
                finalhub_image,
                icon_size=(20, 20)
            )        
        else:
            icon = CustomIcon(
                hub_image,
                icon_size=(10, 10)
            )
        marker=folium.Marker(
            location=[df.iloc[i]['lat'],df.iloc[i]['lng']],
            icon=icon,
            popup=str(df.iloc[i]['ID']),
            tooltip=name
        )
        m.add_child(marker)
```

```python
def coord_list(df):
    ## transform df with 'lat' and 'lng' to location list
    data=[]
    for i in range(len(df)):
        lats=df.iloc[i]['lat']
        lons=df.iloc[i]['lng']
        coor=[lats,lons]
        data.append(coor)

    return data
```

```python
def cal_radius(nums):
    ## nums: input number to decide the radius
    radius_breaks=[0,10,100,1000,10000,100000,1000000]
    radius_stops=np.linspace(2,10,num=7)
    radius=0
    for i in range(len(radius_breaks)):
        
        if nums<radius_breaks[0]:
            radius=radius_stops[0]
            break
        if nums>=radius_breaks[i]:
            radius=radius_stops[i]


    return radius
    
```

```python
def add_circle_markergroup(df,name,r_measure,col,m):
    ## df: with columns['ID','lat','lng']
    ## name: the name of the circle markergroup
    ## r_measure: the column used for measuring the radius of the circle
    ## col: the color of the circle
    ## m: base map
    circlemarker_group=folium.FeatureGroup(name=name)
    
    for i in range(len(df)):
        if r_measure=='':
            r=2
        else:
            r=cal_radius(df.iloc[i][r_measure])  

        folium.CircleMarker(
            location=[df.iloc[i]['lat'],df.iloc[i]['lng']],
            radius=r,
            color=col,
            stroke=False,
            fill=True,
            fill_opacity=1,
            opacity=1,
            popup=str(df.iloc[i]['ID']),
            tooltip=name
        ).add_to(circlemarker_group)
    circlemarker_group.add_to(m)
```

```python
def cal_lanewo(nums):
    ## nums: input number to decide the weights
    weight_breaks=[1,10,100,1000,10000,100000,1000000,10000000]
    weight_stops=np.linspace(1,5,num=8)
    opacity_stops=np.linspace(0.2,1,num=8)
    for i in range(len(weight_breaks)):
        if nums<weight_breaks[0]:
            weight=weight_stops[0]
            opacity=opacity_stops[0]
            break
        if nums>=weight_breaks[i]:
            weight=weight_stops[i]
            opacity=opacity_stops[i]

    return weight,opacity
    
```

```python
def add_lane_group(df,name,w_measure,col,m):
    for i in range(len(df)):
        start=df.iloc[i]['origin'].replace("'","")
        end=df.iloc[i]['next'].replace("'","")
        hops=df.iloc[i]['hop']
        line=[]
        line.append(loc_dic[start])
        line.append(loc_dic[end])
        if w_measure=='':
            w=5
            o=1
        else:
            w,o=cal_lanewo(df.iloc[i][w_measure])
        attr = {'fill': '#2166ac','fond-size':'1'}
        p=folium.PolyLine(
            locations=line,
            weight=w,
            color=col,
            popup=str(df.iloc[i]['hop']),
            opacity=o,
            tooltip=start+" "+end
        ).add_to(m)
        plugins.PolyLineTextPath(
            p,
            '\u2911',
            repeat=True,
            offset=0,
            attributes=attr
        ).add_to(m)
```
